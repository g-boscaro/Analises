```python
nome_arq = "animes.csv"
anime_df = read_csv(nome_arq)
```

```python
#Mostra os 5 primeiros registros do Dataframe
anime_df.head()
```

```python
#Removendo registros duplicados utilizando a coluna UID, se duplicado mantém o primeiro registro encontrado
anime_df = anime_df.drop_duplicates(subset="uid", keep="first")
```

## Ranked

```python
#Criando um novo Dataframe com os dados de UID, Title e Ranked
anime_ranked_df = df({"UID": anime_df["uid"], "Title": anime_df["title"], "Ranked": anime_df["ranked"]})
```

```python
#Ordenando o rank do maior para o menor
anime_ranked_sorted_df = anime_ranked_df.sort_values(by="Ranked")
```

### Achando os registros de maior ranking

```python
#Exibindo os 10 registros de maior ranking
maiores_ranques = anime_ranked_sorted_df.head(10)
maiores_ranques
```

### Achando os registros de menor ranking

```python
#Exibindo os 10 registros de menor ranking
anime_ranked_sorted_df.tail(3) 
```

Se faz necessária a remoção de registros que não possuam valores na coluna Ranked

```python
#Removendo os registros sem valores em qualquer uma das colunas
anime_ranked_menor_df = anime_ranked_sorted_df.dropna()

#Exibindo os 10 registros de menor ranking
menores_ranques = anime_ranked_menor_df.tail(10).sort_values(by="Ranked", ascending = False)
```

### 10 menos ranqueados

```python
#Do menor para o maior
menores_ranques
```

## Popularidade

A

```python
#Criando um novo dataframe com as colunas "uid", "title", "popularity"
anime_pop_df = df({"UID": anime_df["uid"], "Titulo": anime_df["title"], "Popularidade": anime_df["popularity"]})
```

### 10 Mais Populares

### 10 Menos Populares

```python
#Exibindo os 10 registros menos populares
menos_populares = anime_pop_sorted_df.tail(10).sort_values(by="Popularidade", ascending = False)
menos_populares
```

Consideracoes sobre popularidade:

## Score

```python
#Criando dataframe com as colunas "uid", "title", "score"
anime_score_df = df({"UID": anime_df["uid"], "Titulo": anime_df["title"], "Score": anime_df["score"]})
```

```python
#Ordenando dataframe do maior para o menor
anime_score_sorted_df = anime_score_df.sort_values(by = "Score", ascending = False)
anime_score_sorted_df
```

Considerações:

### 10 Menores Scores

```python
#Removendo registros que não possuem valor em qualquer uma das colunas
anime_score_sorted_df = anime_score_sorted_df.dropna()

#Exibindo os 10 registros com menores scores
menores_scores = anime_score_sorted_df.tail(10).sort_values(by="Score")
menores_scores
```

## Membros

```python
#Criando dataframe com as colunas "uid", "title", "members"
anime_members_df = df({"UID": anime_df["uid"], "Titulo": anime_df["title"], "Membros": anime_df["members"]})
anime_members_df
```

## Quantidade de Episódios

```python
#Criando dataframe com as colunas "uid", "title", "episodes"
anime_episodes_df = df({"UID": anime_df["uid"],
                        "Titulo": anime_df["title"],
                        "Qnt. Ep.": anime_df["episodes"],
                        "Exibicao": anime_df["aired"]})
anime_episodes_df
```

```python
#Organizando dataframe pela coluna "Qnt. Ep." de maneira descendente
anime_episodes_sorted_df = anime_episodes_df.sort_values(by = "Qnt. Ep.", ascending = False)
anime_episodes_sorted_df
```

Como o mínimo esperado é que haja ao menos 1 episódio, não faz sentido exibir a menor quantidade de episódios, logo, não há necessidade de remover os registros que estejam com "NaN"

### 10 Animes finalizados com maior quantidade de episódios

```python
mais_episodios = anime_episodes_sorted_df.head(10)
mais_episodios
```
